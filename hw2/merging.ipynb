```python
import re
with open("energy.csv","r") as energy:
    energyDate = []
    energyValue = []
    for line in energy:
        if re.search(r'(\d+)-(\d+)-(\d+).*', line):
            date = re.sub(r'^(\d+)-(\d+)-(\d+).*', r'\1-\2-\3', line)
            energyDate.append(date.rstrip())
            value = re.sub(r'.*,(\d+)$', r'\1', line)
            energyValue.append(value.rstrip())

            
print(energyDate)
print(energyValue)
```

```python
import datetime
energyDate == sorted(energyDate, key=lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
#Yes, the dates are ordered
```

```python
#another method compare two dates:
def compare_dates(date1, date2):
    """return if date1 is earlier than date2
    Format yyyy-mm-dd should be used."""
    
    year1 = re.sub(r'^(\d+)-.*', r'\1', date1)
    year2 = re.sub(r'^(\d+)-.*', r'\1', date2)
    month1 = re.sub(r'.*-(\d+)-.*', r'\1', date1)
    month2 = re.sub(r'.*-(\d+)-.*', r'\1', date2)
    day1 = re.sub(r'.*-(\d+)$', r'\1', date1)
    day2 = re.sub(r'.*-(\d+)$', r'\1', date2)
    if year1 > year2:
        return False
    elif year1 < year2:
        return True
    else:
        if month1 > month2:
            return False
        elif month1 < month2:
            return True
        else:
            if day1 > day2:
                return False
            else:
                return True
            
for i in range(1,(len(energyDate))):
    print(compare_dates(energyDate[i-1], energyDate[i]))
    
#Yes, all the dates are ordered
```

```python
n_energy = 1
currentEnergyDay = energyDate[n_energy-1]
# index stores the 
index = []
```

```python
with open("waterTemperature.csv","r") as temperature:
    for line in temperature:
        if re.search(r'^(\d+).*', line):
            line = line.rstrip()
            #print(line)
            day = re.sub(r'^(\d+),(\d+)/(\d+)/(\d+) (\d+).*', r'20\4-\2-\3', line)
            #print(day)
            #if day is greater than currentEnergyDay, 
            if compare_dates(day, currentEnergyDay) is False:
                index.append(int(re.sub(r'^(\d+),.*', r'\1', line)) - 1)
                n_energy += 1
                currentEnergyDay = energyDate[n_energy-1]

print(index)
```
